```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
X = np.array([[1, 2], [11, 4], [1, 0],
              [10, 2], [10, 4], [10, 0]])
# mencoba akses array X
X[0]
```

```python
# mencoba akses elemen tertentu
X[0,1]
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
# melihat objek kmeans
kmeans
```

```python
label = kmeans.labels_
# melihat objek label
label
```

```python
label[0]
```

```python
clusters = kmeans.fit_predict(X)
#X["label"] = clusters
# melihat objek clusters
clusters
```
